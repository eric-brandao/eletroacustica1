## Aplicando a lei de Kirchhoff das tensões (LKT)

$ V - RI - \left(R + \large\frac{\alpha^{2}}{\text{j}\omega C} \right)$ 

$ V = I\left(R + \large\frac{\alpha^{2}}{\text{j}\omega C} \right)$ 

--------------------------

$ I = \large\frac{V}{R + \frac{\alpha^{2}}{\text{j}\omega C}} $


## A tensão no capacitor é

$ \alpha V_{C} = \alpha^{2} IZ_{C} = \large \frac{\alpha^{2} I}{\text{j}\omega C}$

$ \alpha V_{C} = \large\frac{\alpha^{2} V}{\text{j}\omega C \left(R + \frac{\alpha^{2}}{\text{j}\omega C}\right)}$

$ V_{C} = \large \frac{\alpha V}{\text{j}\omega RC  + \alpha^{2}}$



## Assim, podemos obter a resposta em frequência

$ H_{C} = \large \frac{V_{C}}{V} $

$ H_{C} = \large \frac{\alpha V}{\left(\text{j}\omega RC  + \alpha^{2}\right)V} $

$ H_{C} = \large \frac{\alpha }{\text{j}\omega RC  + \alpha^{2}} $



Para analisar essa equação precisamos definir um espaço de frequências:

$f = [1, 1000]\ \text{Hz}$

Também lembrando que a frequência angular é

$\omega = 2\pi f$

Podemos, ainda, calcular a frequência de corte do filtro:

$\omega_{cut} = \large \frac{\alpha^{2}}{RC}$

$f_{cut} = \large \frac{\alpha^{2}}{2\pi RC} $



```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
R = 1e3  # 1 kOhm
C = 24e-6  # 24 uF
alfa = 3

fcut = alfa**2/(2 * np.pi * R*C)
print(fcut, 'Hz')
```

```python
freqs = np.linspace(1., 1000., 1001) # np.arange(1., 1000.)

fcutIdx = np.where(np.isclose(freqs, fcut, atol=1))[0][1]

omega = lambda f: 2 * np.pi * f
```

```python
Hc = alfa / (1j * omega(freqs) * R * C + alfa**2)

HcMag = np.abs(Hc)
HcAng = 180 * np.angle(Hc) / np.pi

decibel = lambda sinal: 10 * np.log10((sinal**2))
```

```python
fig, axs = plt.subplots(2,1)

axs[0].semilogx(freqs, decibel(HcMag))
axs[0].vlines(fcut, decibel(np.min(HcMag)),
              decibel(np.max(HcMag)/(2**0.5)))
axs[0].hlines(decibel(HcMag[fcutIdx]), np.min(freqs), fcut)

axs[1].semilogx(freqs, HcAng)
axs[1].vlines(fcut, np.min(HcAng), HcAng[fcutIdx])
axs[1].hlines(HcAng[fcutIdx], np.min(freqs), fcut)

fig.tight_layout()
fig.show()
```

## Em aula, encontre a tensão e a FRF sobre o Resistor:

$ I = \large\frac{V}{R + \frac{\alpha^{2}}{\text{j}\omega C}} $

$V_{R} = IR$

$V_{R} = \large\frac{RV}{R + \frac{\alpha^{2}}{\text{j}\omega C}}$

$H_{R} = \large \frac{V_{R}}{V}$

$H_{R} = \large\frac{RV}{V\left(R + \frac{\alpha^{2}}{\text{j}\omega C}\right)}$

$H_{R} = \large\frac{R}{\left(R + \frac{\alpha^{2}}{\text{j}\omega C}\right)}$


Escreva o código necessário para gerar os gráficos da FRF da mesma forma que no exemplo anterior.

```python
Hr = R / (R + alfa**2 / (1j * omega(freqs) * C))
HrMag = np.abs(Hr)
HrAng = 180 * np.angle(Hr) / np.pi
```

```python
fig, axs = plt.subplots(2,1)

axs[0].semilogx(freqs, decibel(HrMag))
axs[0].vlines(fcut, decibel(np.min(HrMag)), decibel(np.max(HrMag)/(2**0.5)))
axs[0].hlines(decibel(HrMag[fcutIdx]), np.min(freqs), fcut)

axs[1].semilogx(freqs, HrAng)
axs[1].vlines(fcut, np.min(HrAng), HrAng[fcutIdx])
axs[1].hlines(HrAng[fcutIdx], np.min(freqs), fcut)

fig.tight_layout()
fig.show()
```
