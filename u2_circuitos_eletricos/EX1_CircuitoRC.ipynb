## Aplicando a lei de Kirchhoff das tensões (LKT)

$ V - RI - \large\frac{\alpha^{2}}{\text{j}\omega C} I = 0$

$ V = I\left(R + \large\frac{\alpha^{2}}{\text{j}\omega C} \right)$ 

\begin{equation}
 I = \large\frac{V}{R + \frac{\alpha^{2}}{\text{j}\omega C}}
 \tag{1}
\end{equation}


## A tensão no capacitor é

$ \alpha V_{C} = \large \frac{\alpha^{2}}{\text{j}\omega C} I$

Aplicando a Eq. (1)

$ \alpha V_{C} = \large\frac{\alpha^{2} V}{\text{j}\omega C \left(R + \frac{\alpha^{2}}{\text{j}\omega C}\right)}$


$ V_{C} = \large \frac{\alpha V}{\text{j}\omega RC  + \alpha^{2}}$



## Assim, podemos obter a resposta em frequência

$ H_{c}(\omega) = \large \frac{V_{C}}{V} $

$ H_{c}(\omega) = \large \frac{\alpha V}{\left(\text{j}\omega RC  + \alpha^{2}\right)V} $

\begin{equation}
 H_{c}(\omega) = \large \frac{\alpha }{\text{j}\omega RC  + \alpha^{2}}
 \tag{2}
\end{equation}




Frequência de corte do filtro:

$\omega_{cut} = \large \frac{\alpha^{2}}{RC}$

$f_{cut} = \large \frac{\alpha^{2}}{2\pi RC} $

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 14})
```

```python
# Parametros
R = 1e3  # 1 kOhm
C = 1e-6  # 1 uF
alfa = 3
fcut = alfa**2/(2 * np.pi * R*C)
print('Frequência de corte é {:.2f} Hz'.format(fcut))

# Frequencias
freqs = np.linspace(1, 10000, 1001)
omega = 2 * np.pi * freqs

# FRF
Hc = alfa / (1j * omega * R * C + alfa**2)

# plot
fig, axs = plt.subplots(2,1, figsize = (7,5))
axs[0].semilogx(freqs, 20*np.log10(np.abs(Hc)), linewidth = 2)
axs[0].grid(linestyle = '--', which='both')
axs[0].set_ylim((-40, 5))
axs[0].set_xlim((1, 10000))
axs[0].set_ylabel(r'$|H(\omega)|$ [dB]')
axs[1].semilogx(freqs, np.rad2deg(np.angle(Hc)), linewidth = 2)
axs[1].grid(linestyle = '--', which='both')
axs[1].set_ylabel(r'$\angle H(\omega)$ [$^{\circ}$] ')
axs[1].set_xlabel(r'Frequência [Hz]')
axs[1].set_xlim((1, 10000))
fig.tight_layout()
```
