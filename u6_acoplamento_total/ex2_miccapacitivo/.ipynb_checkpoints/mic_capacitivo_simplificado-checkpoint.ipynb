# Acoplamento eletroacstico (microfone capacitivo simplificado)

O desenho esquemtico do microfone simplificado :



O circuito eletromecnico do sistema  dado na figura:



Vamos considerar uma presso sonora de entrada e medir a tenso de sada do sensor.

```python
import numpy as np
from scipy import special
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 14})


def ac_compliance(vol, rho0 = 1.21, c0 = 343):
    Ca = vol/(rho0*c0**2)
    return Ca

def zrad_cylinder(a, freq, rho0 = 1.21, c0 = 343):
    # Equivalent circuit
    Ra1 = (0.504 * rho0 * c0) / (np.pi * a**2)
    Ra2 = (rho0 * c0) / (np.pi * a**2)
    Ca1 = (5.44 * a**3) / (rho0 * c0**2)
    Ma1 = (0.1952 * rho0) / a
    # Equivalent impedance
    omega = 2*np.pi*freq
    Zrad_eq=(-(omega**2)*Ma1*Ra1*Ra2*Ca1+1j*omega*Ma1*(Ra1+Ra2))/\
        (Ra1+Ra2-(omega**2)*Ma1*Ra1*Ca1+1j*omega*(Ra1*Ra2*Ca1+Ma1))

    return Ra1, Ra2, Ca1, Ma1, Zrad_eq
```

# Parmetros do microfone e sensibilidade

```python
# air, freq
rho0 = 1.21
c0 = 343
eta = 1.84e-5
freq = np.arange(1, 50e3, 1)
w = 2*np.pi*freq

# Parametros do diafragma
a = 0.25*2.54e-2; # meia polegada
S = np.pi*a**2 # area
E0 = 200 # Polarization voltage
h0 = 20e-6 # Backplate-diaphragm
Tm0 = 3150 # Tenso Mec
rhom = 8300 # Density of diaphragm 
tm = 7e-6 # thickness of diaphragm

Mm = 4*rhom*tm*S/3 # Massa do diafragma
Mad = Mm/S**2 
Cm = (1/(8*np.pi*Tm0))
Cad = Cm*S**2
Rm = 3
Ra1, Ra2, Ca1, Ma1, Zrad_eq = zrad_cylinder(a, 1, rho0 = 1.21, c0 = 343)

# Cavidade
V_A = S * h0
Ca = ac_compliance(V_A)

# Sens
Sens = (E0/(1j*w*h0*S)) * (Zrad_eq + 1/(1j*w*Ca) + 1j*w*Mm/S**2 + Rm/S**2 + 1/(1j*w*Cm*S**2))**(-1)

##############
fig = plt.figure(figsize = (7,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.xaxis.set_tick_params(which='major', size=10, width=2, direction='in', top='on')
ax.xaxis.set_tick_params(which='minor', size=5, width=2, direction='in', top='on')
ax.yaxis.set_tick_params(which='major', size=10, width=2, direction='in', right='on')
ax.yaxis.set_tick_params(which='minor', size=5, width=2, direction='in', right='on')
ax.semilogx(freq, 20*np.log10(np.abs(Sens)), '-r', linewidth = 2,
            label = "Sensibilidade")
ax.set_xlabel('Frequency [Hz]', labelpad=10)
ax.set_ylabel(r'$20 \ \mathrm{log}(|\tilde{V}/\tilde{P}|)$ [dB]', labelpad=10)
#ax.legend(frameon=True, fontsize=16)
ax.grid(b=None, which='both', axis='both',linestyle = '--')
ax.set_xlim((1,freq[-1]))
ax.set_ylim((-60,-50))
ax.set_xticks([1, 10, 100, 1000, 10000, 50000])
ax.set_xticklabels(('1', '10', '100', '1000', '10000', '50000'));
```
