# **Circuito aproximado para impedância de radiação**

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 14})
```

```python
rho = 1.21
c = 343
```

### Baffle infinito

$R_{A1} = \large\frac{0.1404\rho_{0}c_{0}}{\pi a^{2}}$

$R_{A2} = \large\frac{\rho_{0}c_{0}}{\pi a^{2}}$

$C_{A1} = \large\frac{5.94a^{3}}{\rho_{0}c_{0}^{2}}$

$M_{A1} = \large\frac{0.27\rho_{0}}{a}$


```python
def zrad_baffle_infto(a):
    Ra1 = (0.1404 * rho * c) / (np.pi * a * a)
    Ra2 = (rho * c) / (np.pi * a * a)
    Ca1 = (5.94 * a * a * a) / (rho * c * c)
    Ma1 = (0.27 * rho) / a
    return Ra1, Ra2, Ca1, Ma1
```

### Final de duto longo

$R_{A1} = \large\frac{0.504\rho_{0}c_{0}}{\pi a^{2}}$

$R_{A2} = \large\frac{\rho_{0}c_{0}}{\pi a^{2}}$

$C_{A1} = \large\frac{5.44a^{3}}{\rho_{0}c_{0}^{2}}$

$M_{A1} = \large\frac{0.1952\rho_{0}}{a}$


```python
def zrad_piston_baffle(a):
    Ra1 = (0.1404 * rho * c) / (a**2)
    Ra2 = (rho * c)/(np.pi * a**2)
    Ca1 = 5.94 *a**3/(rho * c**2)
    Ma1 = (0.27 * rho) / a
    return Ra1, Ra2, Ca1, Ma1
```

```python
a12pol = (6/2)*(2.54e-2)

Ra1, Ra2, Ca1, Ma1 = zrad_piston_baffle(a12pol)
```
